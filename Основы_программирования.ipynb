```python
#@title ставка
F = int(input("Введите будущее значение, которое вы хотите иметь на вашем счету: "))
r = float(input("Введите годовую процентную ставку: "))
n = int(input("Введите количество лет, в течение которых Вы планируете оставить деньги на счету: "))
P=F/(1+r)**n
print ("Вам потребуется внести: ", P)
```

```python
#@title десятки
number = int(input("Введите число: "))
print ("Десятки:", number//10%10)
```

```python
#@title сумма цифр
number = int(input("Введите трёхзначное число: "))
a = number//100
b = (number//10)%10
c = number%10
summa = a+b+c
print ("Сумма цифр числа: ", summa)
```

```python
#@title руб_коп_пироги
a = int(input("Введите количество рублей: "))
b = int(input("Введите количество копеек: "))
k = int(input("Введите количество пирогов: "))
a = a*k
b = b*k
a = a+b//100
b = b%100
print ("Стоимость пирогов: ", a, "рублей", b, "копеек")
```

```python
#@title ученики
class1 = int(input("Количество учащихся в первом классе: "))
class2 = int(input("Количество учащихся вщ втором классе: "))
class3 = int(input("Количество учащихся в третьем классе: "))
print("Наименьшее количество парт: ", (class1//2 + class2//2 + class3//2) + (class1%2 + class2%2 + class3%2))
```
